in reg code its is not useful for analysing dataset as a datset so we are going to get tears from reg code there are two types one is num and other is string firstly we convert num to years now there is a logic to convrt whe the num is 1 to 50 its going to be first parts of year 2001 to 2050 and if its goint to be 51 to 99 its going to be second part of year 2001 to 2050 and now anphabets are from a to z its going to be before *2001*

```python
car_dataset['reg_code'] = (car_dataset['reg_code'].astype(str).str.strip().str.upper()
        .replace({'': np.nan, 'NAN': np.nan, 'NaN': np.nan}))
```

```python
car_dataset['is_numeric_reg'] = car_dataset['reg_code'].str.fullmatch(r'\d{2}', na=False)
car_dataset['is_letter_reg'] = car_dataset['reg_code'].str.fullmatch(r'[A-Z]', na=False)
```

```python
def age_code_to_year_modern(code):
    try:
        n = int(code)
    except:
        return np.nan

    if 1 <= n <= 49:
        return 2000 + n

    if 51 <= n <= 99:
        return 2000 + (n - 50)

    return np.nan


car_dataset['reg_year'] = np.nan
```

```python
num_mask = car_dataset['is_numeric_reg']
car_dataset.loc[num_mask, 'reg_year'] = car_dataset.loc[num_mask, 'reg_code'].apply(age_code_to_year_modern)

```

```python
prefix_year_map = {
    'A': 1983, 'B': 1984, 'C': 1985, 'D': 1986,
    'E': 1987, 'F': 1988, 'G': 1989, 'H': 1990,
    'J': 1991, 'K': 1992, 'L': 1993, 'M': 1994,
    'N': 1995, 'P': 1996, 'R': 1997, 'S': 1998,
    'T': 1999, 'V': 1999, 'W': 2000, 'X': 2000,
    'Y': 2001
}

letter_mask = car_dataset['is_letter_reg']
car_dataset.loc[letter_mask, 'reg_year'] = car_dataset.loc[letter_mask, 'reg_code'].map(prefix_year_map)
```

```python
car_dataset.loc[car_dataset['reg_year'] > 2020, 'reg_year'] = np.nan
```

```python
REFERENCE_YEAR = 2021
car_dataset['car_age_from_reg'] = REFERENCE_YEAR - car_dataset['reg_year']
car_dataset['reg_missing_flag'] = car_dataset['reg_code'].isna().astype(int)
```

```python
num_mask = car_dataset['is_numeric_reg']
car_dataset.loc[num_mask, 'reg_year'] = (car_dataset.loc[num_mask, 'reg_code'].apply(age_code_to_year_modern).astype(float))
```

feature engineering

1.log price
The core problem with raw price

Car prices in your dataset have:

Many low and mid-priced cars

A few very expensive luxury & classic cars (Ferrari, DB5, Senna, etc.)

This causes:

Strong right skew

Extreme values dominating plots & statistics

Mean becoming misleading
Log transformation:

Compresses very large values

Keeps small values distinguishable

Preserves ordering (expensive cars are still expensive)

```python
# Define luxury price threshold (top 1%)
luxury_threshold = cleaning_dataset['price'].quantile(0.99)

# Flag luxury / collector vehicles
cleaning_dataset['luxury_model_flag'] = (
    cleaning_dataset['price'] >= luxury_threshold
).astype(int)
```

```python
# Median price per model
model_median_price = (
    cleaning_dataset
    .groupby('standard_model')['price']
    .median()
)

# Map back
cleaning_dataset['model_standard_price'] = (
    cleaning_dataset['standard_model'].map(model_median_price)
)

# Price ratio within model
cleaning_dataset['model_price_ratio'] = (
    cleaning_dataset['price'] / cleaning_dataset['model_standard_price']
)
```

```python
# Remove zero or negative prices
cleaning_dataset = cleaning_dataset[
    cleaning_dataset['price'] > 0
]
```

```python
# Log-transform price to reduce extreme skewness
cleaning_dataset['log_price'] = np.log1p(cleaning_dataset['price'])
```
